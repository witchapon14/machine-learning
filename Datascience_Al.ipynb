```
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/sample_data/Advertising.csv')
ds = df.sort_values(['TV'],ascending=True)
x = np.array(ds['TV'])
x=x.reshape(-1,1)
y = np.array(ds['Sales'])
y=y.reshape(-1,1)
model1 = LinearRegression()
model1.fit(x,y)
y_pred = model1.predict(x)
plt.scatter(x,y)
plt.show()
```

```
from sklearn.linear_model import PolynomialFeatures
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/sample_data/Advertising.csv')
ds = df.sort_values(['TV'],ascending=True)
x = np.array(ds['TV'])
x=x.reshape(-1,1)
y = np.array(ds['Sales'])
y=y.reshape(-1,1)
model1 = LinearRegression()
model1.fit(x,y)
y_pred = model1.predict(x)
plt.scatter(x,y)
plt.show()
```

```
plt.style.use('seaborn')
plt.figure(figsize=[7,5],dpi=100)
plt.plot(x,y_pred,c='r',linewidth=2.3)
plt.scatter(x,y)
plt.show()
print(model1.score(x,y))
```

```
#ประเมินmodel
#1.MAE
a=0
for i in range(len(y)):
  a+=abs(y[i][0]-y_pred[i][0])
print("MAE-->",a/len(y))
```
